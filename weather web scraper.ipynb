```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome() 
```

```python
driver = webdriver.Chrome() 
url = "https://www.wunderground.com/history/monthly/us/ca/los-angeles/KHHR/date/2023-11"
driver.get(url)
```

```python
try:
    # Wait for the table to load
    table = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, "//*[contains(@aria-labelledby, 'History days')]"))
    )
    rows = table.find_elements(By.TAG_NAME, "tr")
    print(f"Found {len(rows)} rows in the table.")
except Exception as e:
    print("Error locating table:", e)
    print("Page source for debugging:")
    print(driver.page_source)
```

```python
table = driver.find_element(By.XPATH, "//*[contains(@aria-labelledby, 'History days')]")
rows = table.find_elements(By.TAG_NAME, "tr")
```

```python
for row in rows:
    cells = row.find_elements(By.TAG_NAME, "td")
    print([cell.text.strip() for cell in cells])
```

```python
data = []

for row in rows:
    cells = row.find_elements(By.TAG_NAME, "td")

    print("Row data:", [cell.text.strip() for cell in cells])
    
    if len(cells) >= 13:
        time = cells[0].text.strip()
        temperature = cells[2].text.strip()
        dew_point = cells[5].text.strip()
        humidity = cells[8].text.strip()
        wind_speed = cells[11].text.strip()
        
        print(f"Time: {time}, Temperature: {temperature}, Humidity: {humidity}, Wind Speed: {wind_speed}")
        
        data.append([time, temperature,dew_point, humidity, wind_speed])
```

```python
for entry in data:
    print(entry)
```

```python
df = pd.DataFrame(data, columns=["Time", "Temperature","Dew Point", "Humidity", "Wind Speed"])

# Save the DataFrame to a CSV file
df.to_csv("weather_data.csv", index=False)
print("Data saved to weather_data.csv")
```
